### Visualize and Analyze the Maximum,Minimum and Median income

###  Visualize and Find out the no of days of reduced activity of males and females seperately due to illness

```python
df_new.groupby('gender')['reduced'].mean().plot(kind='barh', color=['blue', 'green'])
plt.title('Average Reduced Activity by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Average Reduced Activity', fontsize=14)
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.show()
```

### Find out the correlation between variables in the given dataset correlation between different variables 

```python
# Convert categorical columns to one-hot encoding
df_encoded = pd.get_dummies(df_new, drop_first=True)

# Generate the correlation heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(df_encoded.corr(), cbar=True, annot=True, cmap='Blues', fmt=".2f", linewidths=0.5, square=True)
plt.title("Correlation Heatmap with One-Hot Encoded Categorical Columns", fontsize=16)
plt.show()
```

### Analyse how the income of a patient effects the no of visits to the hospital

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 10))

sns.scatterplot(data=df_new, x='income', y='visits', 
                 hue='gender', 
                 size='age',    
                 sizes=(20, 200), 
                 alpha=0.7,    
                 palette='deep') 

plt.xlabel('Income', fontsize=14)
plt.ylabel('Number of Visits', fontsize=14)
plt.title('Income vs. Number of Visits', fontsize=16)
plt.legend(title='Gender')  
plt.grid(True)  
plt.show()
```

### Visualize the percentage of people getting government health insurance due to low income, due to old age and and also the percentage of people having private health insurance

```python
gov_low_income = df_new[df_new['income'] < 0.50].shape[0]  # Government insurance due to low income
gov_old_age = df_new[df_new['age'] > 0.65].shape[0]  # Government insurance due to old age (adjust as needed)
private_health = df_new[df_new['private'] == 'yes'].shape[0]  # Private health insurance

# Total number of respondents
total_respondents = df_new.shape[0]

# Calculating percentages
percentages = {
    'Government Health Insurance (Low Income)': (gov_low_income / total_respondents) * 100,
    'Government Health Insurance (Old Age)': (gov_old_age / total_respondents) * 100,
    'Private Health Insurance': (private_health / total_respondents) * 100
}

# Create a DataFrame for plotting
insurance_df = pd.DataFrame(list(percentages.items()), columns=['Insurance Type', 'Percentage'])

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Insurance Type', y='Percentage', data=insurance_df, palette='pastel')

# Adding titles and labels
plt.title('Percentage of People by Type of Health Insurance', fontsize=16)
plt.xlabel('Type of Insurance', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=14)
plt.xticks(rotation=15)  # Rotate x labels for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

### Relation between Health status and Activity reduction using heat map 

```python
contingency_table = pd.crosstab(df_new['health'], df_new['reduced'], margins=False)

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu', cbar=True)

# Adding titles and labels
plt.title('Heatmap of Health Status vs Activity Reduction', fontsize=16)
plt.xlabel('Activity Reduction', fontsize=14)
plt.ylabel('Health Status', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='income', y='reduced', data=df_new, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Income vs Activity Reduction with Regression Line', fontsize=16)
plt.xlabel('Income', fontsize=14)
plt.ylabel('Activity Reduction', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='health', y='reduced', data=df_new, palette='pastel')
plt.title('Activity Reduction Distribution by Health Status', fontsize=16)
plt.xlabel('Health Status', fontsize=14)
plt.ylabel('Activity Reduction', fontsize=14)
plt.xticks([0, 1], ['Unhealthy', 'Healthy'])  # Adjust labels as needed
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='health', data=df_new, palette='pastel')
plt.title('Count of Health Status', fontsize=16)
plt.xlabel('Health Status', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks([0, 1], ['Unhealthy', 'Healthy'])  # Adjust labels as needed
plt.show()
```
